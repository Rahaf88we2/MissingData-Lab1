```python
# no missing salary for male but we find 26 missing value for female. So i will replace the nan salary for female with the mean salary for female.
condition = HR_Missing["Sex"] == "F"
female = HR_Missing[condition]
female
```

```python
female_mean= round(female['Salary'].mean(),2)
female['Salary'].fillna(female_mean, inplace=True)
```
